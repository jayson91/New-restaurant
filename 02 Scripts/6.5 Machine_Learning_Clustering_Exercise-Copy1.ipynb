```python
kmeans = KMeans(n_clusters = 6) 
kmeans.fit(scaled_df)
kmeans.inertia_
```

With this inertia we can analyze that the best clustering would be 4. For better understanding lets look at the WSS or Within Sum of Squares using diagram.

```python
wss=[]
```

```python
for i in range (1, 11):
    KM= KMeans(n_clusters=i)
    KM.fit(scaled_df)
    wss.append(KM.inertia_)
```

```python
wss
```

```python
plt.plot(range(1,11), wss)
```

```python
kmeans = KMeans(n_clusters = 4) 
kmeans.fit(scaled_df)
labels=kmeans.labels_
```

```python
df["Clus_kmeans"]= labels
df.head(5)
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
silhouette_score(scaled_df, labels)
```

Since the silhouettee score is close to 0.5. We can say that the clusters are fairly distinguished

```python
sil_width= silhouette_samples(scaled_df, labels)
```

```python
df["sil_width"]= sil_width
```

```python
df.head(5)
```

```python
silhouette_samples(scaled_df, labels).min()
```

This indicates that there are certain records are probably mapped to incorrect clusters. This may be due to the data that does not necessarily cluster well, so negative values may be unavoidable. 

```python
df["Clus_kmeans"].value_counts()
```

```python
# Plot the clusters for the "Approx Price" and "Rating" variables.

plt.figure(figsize=(12,8))
ax = sns.scatterplot(x=df['rate'], y=df['approx_cost(for two people)'], hue=kmeans.labels_, s=100) 
# Here, you're subsetting `X` for the x and y arguments to avoid using their labels. 
# `hue` takes the value of the attribute `kmeans.labels_`, which is the result of running the k-means algorithm.
# `s` represents the size of the points you want to see in the plot.

ax.grid(False) # This removes the grid from the background.
plt.xlabel('rate') # Label x-axis.
plt.ylabel('approx_cost(for two people)') # Label y-axis.
plt.show()
```

1. We can easily identify for Cluster 1 the pricing is between 100 to 1800 and the rating is close to 1 to 3.4.
2. Cluster 3 which has the most number of counts has ratingbetween 3.5 to 5 and the pricing is again between 100 to 1500.
3. Cluster 0 has very good rating with approx pricing from 1000 to above 6000.
4. Cluster 2 is merging with both cluster 3 and 0 and the rating is very high but the pricing is between 100 to 3000.
